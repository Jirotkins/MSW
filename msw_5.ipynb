```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

#Bisection method
def bisection(f, a0, b0):
    a, b = a0, b0
    eps = 0.001

    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
        return x

#Newton method using derivatives
def newton(f, a0, b0):
    a, b = a0, b0
    delta = 0.001

    def df(x):
        h = 0.001
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew

#Result method (graph)
def plot_result(f, a0, b0, x, title, method):
    print("Newton method") if method=="newton" else print("Bisection method")
    print("root: ", x)
    plt.title(title)
    plt.plot(x, f(x), "bo")
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r") 
    plt.show()
```

```python
#Function 1
print("Function: 3x^3 - 2x^2 + 3x - 1")
func1 = lambda x: 3*x**3 - 2*x**2 + 3*x - 1
a1, b1 = -1, 1
plot_result(func1, a1, b1, newton(func1, a1, b1), title="3x^3 - 2x^2 + 3x - 1", method="newton")
plot_result(func1, a1, b1, bisection(func1, a1, b1), title="3x^3 - 2x^2 + 3x - 1", method="bisection")
timeb1 = timeit.timeit(lambda: bisection(func1, a1, b1), number=100000)
timen1 = timeit.timeit(lambda: newton(func1, a1, b1), number=100000)
print(f"TIME => BISECTION: {timeb1/100000}s, NEWTON: {timen1/100000}s")
```

```python
#Function 2
print("Funtion: 7x^2 + e^(3x) - 3")
func2 = lambda x: 7*x**2 + np.exp(3*x) - 3
a2, b2 = -1, 1
plot_result(func2, a2, b2, newton(func2, a2, b2), title="7x^2 + e^(3x) - 3", method="newton")
plot_result(func2, a2, b2, bisection(func2, a2, b2), title="7x^2 + e^(3x) - 3", method="bisection")
timeb2 = timeit.timeit(lambda: bisection(func2, a2, b2), number=100000)
timen2 = timeit.timeit(lambda: newton(func2, a2, b2), number=100000)
print(f"TIME => BISECTION: {timeb2/100000}s, NEWTON: {timen2/100000}s")
```
